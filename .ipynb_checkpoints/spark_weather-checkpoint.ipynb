```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import regexp_replace
spark = SparkSession.builder.appName("Weather Data").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://noaa-global-hourly-pds.s3.amazonaws.com/2022/72403093738.csv"
spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("72403093738.csv"), sep=",", header=True)
df.show()
```

```python
df.printSchema()
```

```python
#tempurature and dewpoint in celsius +0139,1 = 13.91C
weather_df = df.select("STATION","DATE","CALL_SIGN","TMP","DEW","WND","MA1")
```

```python
weather_df = weather_df.na.drop()
```

```python
weather_df = weather_df.filter((weather_df.TMP != "+9999,9") & (weather_df.CALL_SIGN != "99999"))
```

```python
weather_df = weather_df.withColumn("Temperature C", regexp_replace(weather_df["TMP"], r",.", "").cast("float")/10)
weather_df = weather_df.withColumn("Dew Point C", regexp_replace(weather_df["DEW"], r",.", "").cast("float")/10)
weather_df = weather_df.withColumn("Pressure", regexp_replace(weather_df["MA1"], r",.+", "").cast("float")/10)
weather_df = weather_df.withColumn("Wind Speed m/s", regexp_replace(weather_df["WND"], r"\d+,\d,.,", ""))
weather_df = weather_df.withColumn("Wind Speed m/s", regexp_replace(weather_df["Wind Speed m/s"], r",.", "").cast("float"))
weather_df = weather_df.withColumn("Wind Direction", regexp_replace(weather_df["WND"], r",.+", "").cast("float"))
weather_df = weather_df.drop("TMP", "DEW", "MA1","WND")
```

```python
weather_df.show(150)
```
